```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import RMSprop

from keras.models import Model

import matplotlib.image as mpimg
import random as r
import matplotlib.pyplot as plt
import numpy as np

if tf.test.gpu_device_name():
    print('Default GPU Device: {}'.format(tf.test.gpu_device_name()))
else:
    print('Please install GPU version of TF')

class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('accuracy')>0.99):
      print("\nReached 99% accuracy stopping")
      self.model.stop_training = True               
callbacks = myCallback()

train_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)
train_generator = train_datagen.flow_from_directory(
        'C:/Users/crims/Tensorflow/Practice/RockpaperDataset/rps/',
        target_size=(224, 224),
        color_mode='rgb',
        batch_size = 32,
        class_mode='categorical',
        subset='training',
        shuffle=True)

valid_generator = train_datagen.flow_from_directory(
        'C:/Users/crims/Tensorflow/Practice/RockpaperDataset/rps-test-set/',
        target_size=(224, 224),
        color_mode='rgb',
        batch_size = 32,
        class_mode='categorical',
        subset='validation')
```

```python
#Model
resnet50 = keras.Sequential(
    [ 
        layers.Conv2D(32, (3,3), activation='relu', input_shape=(150,150,3)),
        layers.MaxPooling2D(2,2),
        layers.Conv2D(64, (3,3), activation='relu'),
        layers.MaxPooling2D(2,2),
        layers.Conv2D(128, (3,3), activation='relu'),
        layers.MaxPooling2D(2,2),
        #flatten results to feed into DNN
        layers.Flatten(),
        layers.Dropout(0.5), #trick to imporve efficiency by dropping neurons
        #512 neuron hidden layer

        layers.Dense(512, activation='relu'),
        layers.Dense(3, activation='softmax')
    ]
)

# resnet50 = tf.keras.applications.ResNet50(include_top=True, weights=None, 
#                                           input_shape=(224,224,3), classes=3)

resnet50.compile(loss = 'categorical_crossentropy', optimizer='RMSprop',  metrics=['accuracy'])

resnet50.summary()
```

```python
#Train model
resnet50.fit(train_generator, validation_data=valid_generator,
          epochs=5, batch_size = 16) #, callbacks=[callbacks]))

#Evaluate the model
results = resnet50.evaluate(valid_generator)
resnet50.save('resnet50_Rockpapermodel')
```
